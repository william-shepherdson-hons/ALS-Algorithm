```python
import pandas as pd

input_path = "train_data.csv"

output_path = "train_data.csv"

# Load CSV
df = pd.read_csv(input_path)

# Replace all skill IDs with 0
df['skill_id'] = 0

# Save updated CSV
df.to_csv(output_path, index=False)

print("Updated file saved:", output_path)
```

```python
from pyBKT.models import Model

defaults = {'skill_name': 'skill_id'}
model = Model(seed = 42, num_fits = 1, defaults = defaults)
print("Fitting model")
model.fit(data_path="train_data.csv")
```

```python
print(model.params())
```

```python
preds_df = model.predict(data_path="test_data.csv")
rmse = model.evaluate(data_path="test_data.csv")
auc = model.evaluate(data_path="test_data.csv", metric="auc")
accuracy = model.evaluate(data_path="test_data.csv", metric="accuracy")
print("RMSE: " + str(rmse))
print("AUC: " + str(auc))
print("Accuracy:" + str(accuracy))
print(preds_df)
```
